```python
data_dir = "/kaggle/input/comment-spoiler-classification/"
```

```python
data1 = pd.read_csv(data_dir + 'anime-comments-1.csv', sep='\t', usecols=['Text'])
data2 = pd.read_csv(data_dir + 'anime-comments-2.csv', sep='\t', usecols = ['Text'])
data3 = pd.read_csv(data_dir + 'anime-comments-3.csv', sep='\t', usecols=['Text'])
```

```python
print(data1.shape)
print(data2.shape)
print(data3.shape)
```

```python
data1.head()
```

```python
full_data = pd.concat([data1, data2, data3], axis=0, ignore_index=True)
```

```python
full_data.shape
```

```python
from transformers import pipeline

pipe = pipeline(model="facebook/bart-large-mnli")
```

```python
full_data.head()
```

```python
pipe("I can't believe the write killed off Nanami so early",
    candidate_labels=["spoiler", "friendly", "neutral"]
)

```
