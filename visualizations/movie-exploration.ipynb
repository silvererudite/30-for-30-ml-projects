```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
sns.set_theme(style="whitegrid")
rcParams['figure.figsize'] = 15,8
```

```python
df = pd.read_csv('/kaggle/input/the-movies-dataset/movies_metadata.csv')
df.head()
```

```python
df.columns
```

#### The column `Budget` even though has object dtype, there are both numbers and strings for some reason. We cannot directly convert it to a datatype so we opt for coercing the error of any kind other than numbers to NaN 

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
```

```python
df['release_date'] = pd.to_datetime(df['release_date'],  errors='coerce')
```

```python
ax = sns.distplot(df['budget'], kde=False).set_title("Distribution of Budget over the years")
```
