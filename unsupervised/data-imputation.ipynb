```python
from tqdm import tqdm
```

```python
df = pd.read_csv('/kaggle/input/tabular-playground-series-jun-2022/data.csv')
df.head()
```

```python
def generate_report(data):
    print(data.info())
    print(data.describe())
generate_report(df)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df_copy = df.fillna(df.mean() )
```

```python
df_copy.isnull().sum().sort_values(ascending=False)
```

```python
submission = pd.read_csv('/kaggle/input/tabular-playground-series-jun-2022/sample_submission.csv', index_col='row-col')
submission.head()
```

```python
submission_mean = submission.copy()
```

```python
submission.index
```

```python
for i in tqdm(submission.index):
    row = int(i.split('-')[0])
    col = i.split('-')[1]
    submission_mean.loc[i, 'value'] = df_copy.loc[row, col]
```

```python
submission_mean.head()
```

```python
submission_mean.to_csv("submission.csv")
```
